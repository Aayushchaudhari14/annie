```python
sorted_df=df.sort_values(by='petal.length',ascending=False)
sorted_df[['petal.length','variety','sepal.width']].head()
```

```python
transposed_df=df.transpose()
transposed_df
```

```python
melted_df = pd.melt(df,
                 id_vars=['variety'],
                 value_vars=['sepal.width','sepal.length'],
                 var_name='Engagement Type',
                 value_name='Count')
melted_df
```

```python
wide_df=melted_df.pivot_table(index='variety',
                             columns='Engagement Type',
                             values='Count',
                             aggfunc='mean')
wide_df
```
