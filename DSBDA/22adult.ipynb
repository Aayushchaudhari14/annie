```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('adult_dataset.csv')
```

```python
df.head()
```

```python
df.hist(figsize=(15,10),color='skyblue',edgecolor='black')
plt.suptitle("Histogram of cesus bureau dataset", fontsize=16)
plt.show()
```

```python
target_counts=df['income'].value_counts()
labels=['<=50k','>50k']
colors=['mediumseagreen','lightcoral']

plt.figure(figsize=(6,6))
plt.pie(target_counts,labels=labels,autopct='%1.1f%%',startangle=140,colors=colors)
plt.title('Income Distribution')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(x='income',y='hours-per-week',data=df,palette='pastel')
plt.title('Hours per Week by Income Group')
plt.xlabel('income')
plt.ylabel('hours-per-week')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(data=df, x='age', y='hours-per-week', hue='income', palette='Set1')
plt.title('Scatter Plot: Age vs hours-per-week')
plt.xlabel('Age')
plt.ylabel('hours-per-week')
plt.show()
```
