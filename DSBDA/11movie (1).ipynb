```python
melted_df = pd.melt(df,
                    id_vars=['original_language'],
                    value_vars=['vote_average','vote_count'],
                    var_name='Engagement Type',
                    value_name='Count')

melted_df
```

```python
wide_df = melted_df.pivot_table(index='original_language',
                                 columns='Engagement Type',
                                 values='Count',
                                 aggfunc='mean')  # Aggregating using mean for simplicity

wide_df
```
