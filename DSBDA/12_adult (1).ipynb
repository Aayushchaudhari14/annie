```python
melted_df = pd.melt(df,
                    id_vars=['educational-num'],
                    value_vars=['fnlwgt', 'capital-gain'],
                    var_name='Engagement Type',
                    value_name='Count')

melted_df
```

```python
wide_df = melted_df.pivot_table(index='educational-num',
                                 columns='Engagement Type',
                                 values='Count',
                                 aggfunc='mean')  # Aggregating using mean for simplicity

wide_df
```
