```python
import pandas as pd
```

```python
df=pd.read_csv('forestfires.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
def categorize_area(area):
    if area==0:
        return "NotAffected"
    elif area<=5:
        return "PartiallyAffected"
    else:
        return "MostlyAffected"
df['AffectedCategory']=df['area'].apply(categorize_area)
```

```python
subset_loc = df.loc[0:3, ['AffectedCategory','X','Y']] # only the 'AffectedCategory' column
subset_loc
```

```python
subset_iloc=df.iloc[0:5,0:5]
subset_iloc
```

```python
subset_indexing = df[['AffectedCategory']] # All rows selecte here

subset_indexing.head()
```

```python
not_affected = df[df['AffectedCategory'] == "NotAffected"]
partially_affected = df[df['AffectedCategory'] == "PartiallyAffected"]
mostly_affected = df[df['AffectedCategory'] == "MostlyAffected"]
```

```python
merged_subset=pd.concat([partially_affected, mostly_affected],ignore_index=True)
merged_subset.head()
```

```python
melted_df=pd.melt(df,
                  id_vars=['AffectedCategory'],
                  value_vars=['wind','rain'],
                  var_name='Engagement Type',
                  value_name='Count')
melted_df
```

```python
wide_df=melted_df.pivot_table(index='AffectedCategory',
                              columns='Engagement Type',
                              values='Count',
                              aggfunc='mean')
wide_df
```
