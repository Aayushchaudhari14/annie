```python
melted_df=pd.melt(df,
                  id_vars='sex',
                  value_vars=['class','age'],
                  var_name='Engagement Type',
                  value_name='Count')
melted_df
```

```python
wide_df=melted_df.pivot_table(index='sex',
                             columns='Engagement Type',
                             values='Count',
                             aggfunc='mean')
wide_df
```
