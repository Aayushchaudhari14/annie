```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('adult_dataset.csv')
```

```python
df.head()
```

```python
df.hist(figsize=(15,10),color='skyblue',edgecolor='black')
plt.suptitle("Histogram of cesus bureau dataset", fontsize=16)
plt.show()
```

```python
sns.stripplot(data=df,x='hours-per-week',jitter=True)
plt.title('Dot Plot of Hours per Week by Education')
plt.xlabel('education')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='gender', palette='coolwarm')
plt.title('Bar Plot of gender')
plt.xlabel('gender')
plt.ylabel('No of Male and Female ')
plt.show()
```

```python
df_sorted=df.sort_values(by='age')
plt.figure(figsize=(10,6))
plt.plot(df_sorted['age'],df_sorted['educational-num'],marker='o',color='orange')
plt.title('Line Chart of Age vs educational-num')
plt.xlabel=('Age')
plt.ylabel=('educational-num')
plt.grid(True)
plt.show()
```

```python
fig,ax=plt.subplots(3,1,figsize=(8,9),gridspec_kw={'height_ratios':[1,2,2]})

sns.boxplot(data=df,x='age',ax=ax[0],color='lightblue')
ax[0].set_title('Boxplot of Max Heart Rate')

sns.histplot(df['age'],bins=20,kde=True,ax=ax[1],color='salmon')
ax[1].set_title('Histogram of Max Heart Rate')

sns.scatterplot(data=df,x='age',y='hours-per-week',palette='Set1',ax=ax[2])
ax[2].set_title('Scatter Plot: Age vs Max Heart Rate(Colored by Target)')

plt.tight_layout()
plt.show()
```
