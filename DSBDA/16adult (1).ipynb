```python
le = LabelEncoder()
for column in df.columns:
    if df[column].dtype == 'object' or df[column].dtype.name == 'category':
        df[column] = le.fit_transform(df[column].astype(str))

# Print the head of the dataframe
print(df.head())
```

```python
for col in df.columns:
    if df[col].min() < 0:
        df = df[df[col] >= 0]

df.head(15)
```

```python
df.dropna(inplace=True)

df.head()
```

```python
cols=['age','capital-gain','capital-loss','hours-per-week']
Q1 = df[cols].quantile(0.25)   # 25th percentile
Q3 = df[cols].quantile(0.75)   # 75th percentile
IQR = Q3 - Q1            # Interquartile Range (middle 50% of data)

# Outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove rows where any column has a value outside the IQR range
df = df[~((df[cols]< lower_bound) | (df[cols]> upper_bound)).any(axis=1)]

df.head()
```
